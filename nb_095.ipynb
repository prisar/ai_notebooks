## Video processing and chunking

### Subtask:
Split the video into 2-5 minute chunks and extract audio from each chunk.


**Reasoning**:
The subtask requires video processing, which can be done using the moviepy library. I will install the library first.

**Reasoning**:
The moviepy library is installed. Now I need to load the video, determine its duration, calculate chunk timestamps, extract chunks and their audio, and save the audio files. I will combine these steps into a single code block.



```python
from moviepy.editor import VideoFileClip
import os

# Define the video file path (replace with your video file)
video_path = '/tmp/interview_video.mp4'
output_dir = '/tmp/audio_chunks'

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Load the video clip
try:
    video_clip = VideoFileClip(video_path)
except Exception as e:
    print(f"Error loading video file: {e}")
    # Finish the task with failure if the video file cannot be loaded
    raise

# Get the total duration in seconds
total_duration = video_clip.duration
print(f"Total video duration: {total_duration} seconds")

# Define chunk length range in seconds (2 to 5 minutes)
min_chunk_length = 2 * 60
max_chunk_length = 5 * 60

# Calculate chunk timestamps
start_time = 0
chunk_index = 0
timestamps = []

while start_time < total_duration:
    end_time = min(start_time + max_chunk_length, total_duration)
    if end_time - start_time < min_chunk_length and end_time < total_duration:
      end_time = min(start_time + min_chunk_length, total_duration)

    timestamps.append((start_time, end_time))
    start_time = end_time
    chunk_index += 1

print(f"Calculated {len(timestamps)} chunks.")

# Extract audio from each chunk and save
for i, (start, end) in enumerate(timestamps):
    try:
        # Extract the subclip for the current chunk
        chunk_clip = video_clip.subclip(start, end)

        # Extract the audio stream
        audio_clip = chunk_clip.audio

        # Define the output audio file path
        audio_output_path = os.path.join(output_dir, f'audio_chunk_{i+1}.wav')

        # Write the audio to a WAV file
        audio_clip.write_audiofile(audio_output_path, codec='pcm_s16le')

        print(f"Extracted and saved audio chunk {i+1} ({start:.2f}s to {end:.2f}s) to {audio_output_path}")

        # Close the audio clip to release resources
        audio_clip.close()
        chunk_clip.close()

    except Exception as e:
        print(f"Error processing chunk {i+1}: {e}")
        # Continue to the next chunk even if one fails

# Close the main video clip
video_clip.close()

print("Audio extraction complete.")
```

## Vector encoding

### Subtask:
Encode the transcriptions into vector representations using a suitable model.


**Reasoning**:
Import the necessary libraries and a suitable pre-trained model and its tokenizer for generating vector embeddings.



```python
from transformers import AutoTokenizer, AutoModel
import torch
import os

# Load a pre-trained model and tokenizer (e.g., Sentence-BERT)
# Using 'sentence-transformers/all-MiniLM-L6-v2' as an example
tokenizer = AutoTokenizer.from_pretrained('sentence-transformers/all-MiniLM-L6-v2')
model = AutoModel.from_pretrained('sentence-transformers/all-MiniLM-L6-v2')

# Function to get embeddings for a text
def get_embeddings(text):
    # Tokenize the text
    encoded_input = tokenizer(text, padding=True, truncation=True, return_tensors='pt')

    # Compute token embeddings
    with torch.no_grad():
        model_output = model(**encoded_input)

    # Perform pooling. In this case, mean pooling.
    # Get the embeddings for the [CLS] token (usually the first token) or mean pooling
    # Using mean pooling as it's common for Sentence-BERT
    input_mask_expanded = encoded_input['attention_mask'].unsqueeze(-1).expand(model_output.last_hidden_state.size()).float()
    sum_embeddings = torch.sum(model_output.last_hidden_state * input_mask_expanded, 1)
    sum_mask = torch.clamp(input_mask_expanded.sum(1), min=1e-9)
    mean_embeddings = sum_embeddings / sum_mask

    return mean_embeddings.squeeze().tolist() # Return as a list

# Dictionary to store vector embeddings
audio_embeddings = {}

# Check if audio_transcriptions is empty or not available
if not audio_transcriptions:
    print("audio_transcriptions is empty or not available. Cannot generate embeddings.")
else:
    # Iterate through transcriptions and generate embeddings
    for filename, transcription in audio_transcriptions.items():
        if transcription and isinstance(transcription, str): # Ensure transcription is not empty and is a string
            try:
                print(f"Generating embedding for {filename}...")
                embedding = get_embeddings(transcription)
                audio_embeddings[filename] = embedding
                print(f"Embedding generated for {filename}.")
            except Exception as e:
                print(f"Error generating embedding for {filename}: {e}")
        else:
            print(f"Skipping {filename} due to empty or invalid transcription.")

# Print the generated embeddings (optional, might be large)
# print("\n--- Audio Embeddings ---")
# for filename, embedding in audio_embeddings.items():
#     print(f"{filename}: {embedding[:10]}...") # Print only first 10 elements to avoid flooding output

# No dataframes are used in this step.
```
